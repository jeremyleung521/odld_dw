# Plot Potential

```python
import numpy as np
import matplotlib.pyplot as plt
A = 2
B = 20
C = 0
sigma = 0.0001 ** (0.5)
gradfactor = sigma * sigma / 2
reflect_at = 10
reflect_zero = 0
```

```python
def odld_1d_potential(x, A=2, B=20, C=0, normalize=True):
    potential = A**2 / np.sin(x)**2 + (-1/2) *  B * np.cos(2 * x) - (0.25) / np.sin(x)**2

    # Needed to normalize so Minimum lies at 0
    if normalize is True:
        potential -= np.min(potential)

    return potential
```

```python
x = np.linspace(0.001, np.pi-0.001, 1000)
y = odld_1d_potential(x, A=A, B=B, C=C, normalize=False)
plt.rcParams['figure.figsize'] = [8,6.5]
plt.rcParams['font.size'] = 20
plt.plot(x, y)
plt.plot(np.pi/4, odld_1d_potential(np.pi/4, A=A, B=B, C=C, normalize=False),marker='*',color='green',markersize=30)
plt.axvline(np.pi,color='red')
plt.axvline(0,color='red')
plt.ylabel('energy')
plt.xlabel('x-coordinate')
plt.ylim(5, 20)
plt.xlim(-np.pi / 8, (9/8) * np.pi)
plt.xticks(ticks=[0, np.pi/2, np.pi], labels=[0, r'$\pi/2$', r'$\pi$'])
plt.xticks(ticks=[np.pi/4, (3/4) * np.pi], labels=[r'$\pi/4$', r'$\dfrac{3\pi}{4}$'], minor=True)
plt.text(np.pi/32,15,'<-         reflecting barriers        ->', color='red')
plt.text(2* np.pi/16,6,'basis state', color='green')
plt.title('Absolute energy');
```

```python
x = np.linspace(0.001, np.pi-0.001, 1000)
y = odld_1d_potential(x, A=A, B=B, C=C, normalize=False)
min = np.min(y)
y -= min
plt.rcParams['figure.figsize'] = [8,6.5]
plt.rcParams['font.size'] = 20
plt.plot(x, y)
plt.plot(np.pi/4, odld_1d_potential(np.pi/4, A=A, B=B, C=C, normalize=False)/ min,marker='*',color='green',markersize=30)
plt.axvline(np.pi,color='red')
plt.axvline(0,color='red')
plt.ylabel('relative energy (to minimum)')
plt.xlabel('x-coordinate')
plt.ylim(-2, 15)
plt.xlim(-np.pi / 8, (9/8) * np.pi)
plt.xticks(ticks=[0, np.pi/2, np.pi], labels=[0, r'$\pi/2$', r'$\pi$'])
plt.xticks(ticks=[np.pi/4, (3/4) * np.pi], labels=[r'$\pi/4$', r'$\dfrac{3\pi}{4}$'], minor=True)
plt.text(np.pi/32,10,'<-         reflecting barriers        ->', color='red')
plt.text(2* np.pi/16,-1.5,'basis state', color='green')
plt.title('Normalized energy (energy minimum is 0)');
```
